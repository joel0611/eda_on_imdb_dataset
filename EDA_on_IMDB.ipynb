### Data Cleaning

```python
df_cleaned = df.drop(['ID', 'Poster_Link', 'Overview'], axis=1)
```

```python
df_cleaned['Runtime'] = df_cleaned['Runtime'].str.replace(' min', '').astype(int)
```

```python
df_cleaned['Gross'] = df_cleaned['Gross'].str.replace(',', '', regex=False).astype(float)
```

```python
df_cleaned.loc[df_cleaned['Released_Year'] == 'PG', 'Released_Year'] = '1975'
```

```python
df_cleaned['Released_Year'] = df_cleaned['Released_Year'].astype(int)
```

```python
df_cleaned['Primary_Genre'] = df_cleaned['Genre'].apply(lambda x: x.split(',')[0])
```

### Data Visualization

```python
# 1. Correlation Heatmap for Numerical Features
plt.figure(figsize=(10, 8))
sns.heatmap(df_cleaned[['IMDB_Rating', 'Meta_score', 'No_of_Votes', 'Gross', 'Runtime', 'Released_Year']].corr(), annot=True, cmap='viridis', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Features')
plt.tight_layout()
plt.savefig('correlation_heatmap.png')
print("\nGenerated correlation_heatmap.png")
```

```python
# 2. Histograms for Distribution of Numerical Features
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('Histograms of Numerical Features')
sns.histplot(df_cleaned['IMDB_Rating'], ax=axes[0, 0], kde=True)
axes[0, 0].set_title('IMDB Rating Distribution')
sns.histplot(df_cleaned['Meta_score'], ax=axes[0, 1], kde=True)
axes[0, 1].set_title('Meta Score Distribution')
sns.histplot(df_cleaned['Runtime'], ax=axes[1, 0], kde=True)
axes[1, 0].set_title('Runtime Distribution')
sns.histplot(df_cleaned['Gross'], ax=axes[1, 1], kde=True)
axes[1, 1].set_title('Gross Revenue Distribution')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig('histograms.png')
print("Generated histograms.png")
```

```python
# 3. Bar plot for the number of movies in each Primary Genre
plt.figure(figsize=(12, 8))
genre_counts = df_cleaned['Primary_Genre'].value_counts()
sns.barplot(x=genre_counts.values, y=genre_counts.index, hue=genre_counts.index, palette='mako', dodge=False, legend=True)
plt.title('Number of Movies per Primary Genre')
plt.xlabel('Number of Movies')
plt.ylabel('Primary Genre')
plt.tight_layout()
plt.savefig('genre_counts.png')
print("Generated genre_counts.png")
```

```python
# 4. Bar plot for the average IMDB rating for each Primary Genre
plt.figure(figsize=(12, 8))
genre_ratings = df_cleaned.groupby('Primary_Genre')['IMDB_Rating'].mean().sort_values(ascending=False)
sns.barplot(x=genre_ratings.values, y=genre_ratings.index, hue=genre_ratings.index, palette='rocket', dodge=False, legend=False)
plt.title('Average IMDB Rating per Primary Genre')
plt.xlabel('Average IMDB Rating')
plt.ylabel('Primary Genre')
plt.tight_layout()
plt.savefig('genre_ratings.png')
print("Generated genre_ratings.png")
```

```python
# 5. Scatter plot of IMDB Rating vs. Gross Revenue
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_cleaned, x='IMDB_Rating', y='Gross')
plt.title('IMDB Rating vs. Gross Revenue')
plt.xlabel('IMDB Rating')
plt.ylabel('Gross Revenue')
plt.tight_layout()
plt.savefig('rating_vs_gross.png')
print("Generated rating_vs_gross.png")
```

```python
# 6. Bar plot for the Top 10 Directors by Average IMDB Rating
plt.figure(figsize=(12, 8))
top_directors = df_cleaned.groupby('Director')['IMDB_Rating'].mean().nlargest(10).sort_values(ascending=False)
sns.barplot(x=top_directors.values, y=top_directors.index, hue=top_directors.index, palette='viridis', dodge=False, legend=False)
plt.title('Top 10 Directors by Average IMDB Rating')
plt.xlabel('Average IMDB Rating')
plt.ylabel('Director')
plt.tight_layout()
plt.savefig('top_directors.png')
print("Generated top_directors.png")
```

```python
# 7. Histogram for the distribution of movies by their release year
plt.figure(figsize=(12, 6))
sns.histplot(data=df_cleaned, x='Released_Year', bins=30, kde=True)
plt.title('Distribution of Movies by Released Year')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.tight_layout()
plt.savefig('movies_by_year.png')
print("Generated movies_by_year.png")
```

### **Insights from the Analysis**

1. **Correlation of Numerical Features**\
The correlation heatmap reveals the relationships between the numerical variables in the dataset.


*   There is a moderate positive correlation of 0.61 between `No_of_Votes` and `Gross revenue`, which is expected as popular movies (more votes) tend to earn more.
*   `IMDB_Rating` has a weak positive correlation with `Gross revenue` (0.15) and `Meta_score` (0.27).
* `Runtime` does not show a strong correlation with any other variable.

2. **Distribution of Numerical Features**\
The histograms show the distribution of key numerical data points:

* **IMDB Rating**: The ratings are left-skewed, with most movies having a rating between 7.5 and 8.5.
* **Meta Score**: The metascores are also left-skewed, with a large number of movies having scores between 70 and 90.
* **Runtime**: Most movies have a runtime between 100 and 140 minutes.
* **Gross Revenue**: The distribution of gross revenue is heavily right-skewed, indicating that a few movies have exceptionally high earnings compared to the majority.

3. **Genre Analysis**\
* **Most Common Genres**: *Drama* is the most common primary genre in this dataset, followed by *Action* and *Comedy*.

* **Top Rated Genres**: The highest average IMDB ratings belong to the *Animation*, *Biography*, and *Western genres*.

4. **IMDB Rating vs. Gross Revenue**\
The scatter plot of IMDB rating versus gross revenue does not show a strong linear relationship. While some highly-rated movies have high gross revenue, there are also many highly-rated movies with lower revenues and vice-versa. This suggests that a high IMDB rating alone does not guarantee high box office success.

5. **Top Directors**\
The analysis of directors shows that certain directors consistently produce highly-rated films. The top 10 directors with the highest average IMDB ratings in this dataset are led by Frank Darabont, followed by Christopher Nolan and Stanley Kubrick.

6. **Movie Releases Over the Years**\
The number of movies in this dataset has been increasing over the years, with a significant peak in the 2000s and 2010s. This reflects the general trend of increasing movie production over time.
